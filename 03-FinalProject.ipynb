## Abstract

This project investigates how different styles of social media use are associated with adult wellbeing. We focus on the distinction between passive consumption, such as browsing feeds and watching short-form content, and active interaction, such as posting and other visible engagement behaviors. Our primary outcomes are self-reported stress and happiness among adults aged 18+, and we also use supporting datasets to examine related outcomes such as sleep issues, depressed mood, and broader mental wellness proxies.

Using a harmonized multi-dataset association framework, our main analysis centers on a large Instagram activity and lifestyle dataset that directly includes a passive-use proxy (`time_on_feed_per_day`), an active-use proxy (`posts_created_per_week`), the primary outcomes, and a broad set of demographic and lifestyle controls. We complement that primary dataset with four supporting datasets that capture broader screen-time exposure, social-comparison behaviors, emotion labels, and wellness-related outcomes. Across these analyses, passive use shows the clearest negative association with wellbeing, with higher passive exposure tracking more strongly with higher stress and lower happiness than active posting behavior. The active-use results are weaker and more mixed than our original hypothesis predicted. Because all datasets are observational and largely self-reported, our conclusions are correlational rather than causal, and we discuss limitations related to measurement choices, sample bias, and external validity.

## Research Question

How are different styles of social media use, passive consumption versus active interaction, associated with wellbeing outcomes among adults aged 18+? In this project, we define passive consumption as behaviors such as browsing feeds and watching short-form content, and we define active interaction as behaviors such as posting and other visible engagement actions. Our primary outcomes are self-reported stress and self-reported happiness, and we use supporting analyses to examine related outcomes such as sleep issues and broader mental-health proxies when those are the closest available measures. Because this is an observational study, we focus on associations rather than causal claims. In the primary analysis, we account for key demographic, socioeconomic, and lifestyle controls including age, gender, country, urban-rural setting, income level, employment status, education level, exercise, sleep, and weekly work hours; in supporting datasets, we use the closest available control variables.

## Background and Prior Work

Social media use is now a major component of everyday life. Instagram, Facebook, TikTok, and Snapchat have accrued daily engagement from a large share of adults. As social media use has grown, research has shifted toward focusing on the manner in which individuals engage with it. For this study, we use a widely adopted distinction in which active interaction includes posting content, commenting, liking, and direct messaging, while passive consumption includes browsing feeds or watching short-form videos. Our initial research suggests that these usage styles may influence different psychological mechanisms, with passive use often linked to social comparison and envy, while active use may have higher perceived support and social connectedness. However, associations with other wellbeing outcomes, including stress, happiness, sleep quality, and physical health, remain mixed and motivate further investigation.

One prior influential effort in this area is the Social Media Activity Questionnaire (SMAQ), which surveyed 1,230 participants derived from the Facebook Activity Questionnaire. This study by Ozimek, Brailovskaia, and Bierhoff separated active and passive engagements with a rating scale from 1 to 5.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) With the application of exploratory factor analysis (EFA) to identify behavioral dimensions, outcomes regarding mental health were assessed using established scales. These include the Bergen Social Media Addiction Scale, the Fear of Missing Out (FoMO) scale, and the Depression Anxiety Stress Scales (DASS). In contrast to our project, this study focused on negative mental-health indicators rather than trying to estimate wellbeing directly, which remains our priority. Their results showed stronger associations between active use and depression, anxiety, and stress, while passive use was more strongly associated with problematic behavioral tendencies such as addiction and FoMO. The authors attribute these tendencies in part to upward social comparison and envy enabled by consumed content.

One notable limitation of the SMAQ study was the disproportionate sample of mostly young and female participants, which limits generalizability to the broader population. In addition, the dataset was limited to Facebook users and geographically restricted to the Ruhr region of Germany. This makes it difficult to extrapolate those observations to older adults, different cultural contexts, and our specific platform focus on Instagram.

Another referenced perspective is provided by the study “Are active and passive social media use related to mental health, wellbeing, and social support outcomes?” This meta-analysis synthesizes findings from 141 quantitative studies that examine correlations between social media use styles and 13 mental-health and wellbeing outcomes. Using pooled effect sizes, Godard and Holtzman report that neither active nor passive use strongly predicts wellbeing on average, with most effects being quite small.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Active use has a modest association with perceived social support and wellbeing, but also with slightly higher anxiety. Passive use shows overall weak associations but is more linked to worse emotional outcomes in general contexts. Crucially, this study highlights that demographic characteristics and contextual moderators, such as age and usage setting, have more substantial influence on outcomes. These findings show the importance of controlling for confounding factors rather than relying on simplistic narratives like “active is good, passive is bad,” which directly informs our project design.

A study that strongly grounds age as a contextual moderator is Underwood’s expanded literature review on adolescents aged 10 to 19. Reviewing 16 peer-reviewed studies gathered from MEDLINE and PubMed via keyword search, the review concludes that both active and passive social media use are positively associated with negative mental-health outcomes in adolescents.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) While the evidence was insufficient to conclude that one style is categorically more harmful, this study reinforces the importance of grounding research by age because the observed effects varied across developmental stages. Our current study responds to that issue by narrowing the analysis to Instagram-related measures among adults aged 18+. We also examine both positive and negative wellbeing outcomes, stretching beyond happiness and stress to include related health indicators such as sleep while explicitly accounting for demographic, socioeconomic, and lifestyle confounders.

## Hypothesis


We hypothesize that among adults aged 18+, higher passive social media consumption will be associated with higher self-reported stress and lower self-reported happiness, while higher active interaction will be associated with lower stress and higher happiness. This hypothesis follows from the idea that passive browsing is more likely to intensify social comparison and rumination, whereas active interaction is more likely to involve social connection and perceived support, which may buffer stress and improve mood.

### Dataset #1: Instagram User Activity and Wellbeing

This dataset is the main analysis source for our project because it is the clearest direct match to our research question. It contains user-level Instagram activity, self-reported wellbeing outcomes, and a useful set of demographic and lifestyle controls. In our current workflow, the raw dataset contains **1,547,896 rows** and **58 columns**; after restricting the analysis to adults aged 18+ and keeping the variables most relevant to our project, the processed analysis table contains **1,401,638 rows** and **14 selected columns**.

The most important variables in this dataset are `time_on_feed_per_day` as our passive-use proxy, `posts_created_per_week` as our active-use proxy, `perceived_stress_score` as our stress outcome, and `self_reported_happiness` as our happiness outcome. We also retain controls for age, gender, country, urban/rural setting, income, employment, education, exercise, sleep, and weekly work hours. These variables make Dataset #1 the strongest available source for directly comparing passive and active social media behavior against our primary wellbeing outcomes.

#### Data wrangling and cleaning decisions
- Load the raw dataset, inspect shape and available columns, and confirm the required analysis variables are present.
- Convert the core analysis fields to numeric where needed, especially age, passive-use, active-use, stress, and happiness.
- Restrict the sample to **adults aged 18+**.
- Keep the selected predictors, outcomes, and controls for analysis instead of carrying all raw columns forward.
- Save the adult-only subset to `data/01-interim/social_media_user_activity_interim.csv` and `data/02-processed/social_media_user_activity_cleaned.csv` for reproducibility.
- Check duplicate rows and missingness in the retained columns before building the EDA-ready table.
- Build `df1_eda` by dropping rows missing the four key analysis variables: passive use, active use, stress, and happiness.
- In this run, missingness in the retained variables is effectively zero, so no imputation step was needed.

#### Potential concerns and biases
- **Cross-sectional observational structure:** We cannot make causal claims; results are associational only.
- **Measurement/unit mismatch:** Passive use is measured in minutes per day, while active use is measured in posts per week, so direct effect-size comparison requires care.
- **External validity concern:** Near-zero missingness across many fields suggests the dataset may be synthetic or heavily processed.
- **Residual confounding:** Even with controls, unmeasured factors may still influence the observed associations.
- **Operationalization limitation:** Passive and active use are each represented by one main proxy variable rather than a richer composite measure.

```python
from pathlib import Path
import pandas as pd
from IPython.display import display, Markdown

RAW_DIR = Path("data/00-raw")
INTERIM_DIR = Path("data/01-interim")
PROCESSED_DIR = Path("data/02-processed")

INTERIM_DIR.mkdir(parents=True, exist_ok=True)
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

processed_path = PROCESSED_DIR / "social_media_user_activity_cleaned.csv"
interim_path = INTERIM_DIR / "social_media_user_activity_interim.csv"
raw_path = RAW_DIR / "instagram_usage_lifestyle.csv"

keep_cols = [
    "age",
    "time_on_feed_per_day",
    "posts_created_per_week",
    "perceived_stress_score",
    "self_reported_happiness",
    "gender",
    "country",
    "urban_rural",
    "income_level",
    "employment_status",
    "education_level",
    "exercise_hours_per_week",
    "sleep_hours_per_night",
    "weekly_work_hours",
]

required_cols = [
    "age",
    "time_on_feed_per_day",
    "posts_created_per_week",
    "perceived_stress_score",
    "self_reported_happiness",
]

display(Markdown("### Dataset #1 loaded and checked"))

df1 = None
source_used = None

if processed_path.exists():
    df1 = pd.read_csv(processed_path)
    source_used = processed_path
elif raw_path.exists():
    df_raw = pd.read_csv(raw_path)
    df_raw["age"] = pd.to_numeric(df_raw["age"], errors="coerce")

    keep_cols_present = [c for c in keep_cols if c in df_raw.columns]
    df1 = df_raw[df_raw["age"].notna() & (df_raw["age"] >= 18)].copy()
    df1 = df1[keep_cols_present]

    df1.to_csv(interim_path, index=False)
    df1.to_csv(processed_path, index=False)
    source_used = raw_path

    print("Built interim dataset:", interim_path)
    print("Built processed dataset:", processed_path)
else:
    raise FileNotFoundError(
        "Dataset #1 was not found. Expected `data/00-raw/instagram_usage_lifestyle.csv` "
        "or `data/02-processed/social_media_user_activity_cleaned.csv`."
    )

print("Source:", source_used)
print("Shape:", df1.shape)
print("Columns:", list(df1.columns))

missing_required = [c for c in required_cols if c not in df1.columns]
if missing_required:
    raise ValueError(f"Required columns missing: {missing_required}")

for col in required_cols:
    df1[col] = pd.to_numeric(df1[col], errors="coerce")

print("\nDuplicate rows:", int(df1.duplicated().sum()))
print("Age min / max:", float(df1["age"].min()), float(df1["age"].max()))

missing_tbl = (
    df1.isna()
       .sum()
       .to_frame("missing_count")
       .assign(missing_pct=lambda x: (x["missing_count"] / len(df1) * 100).round(2))
       .sort_values("missing_count", ascending=False)
)

display(Markdown("### Missingness overview (top columns)"))
display(missing_tbl.head(15))

df1_eda = df1.dropna(
    subset=[
        "time_on_feed_per_day",
        "posts_created_per_week",
        "perceived_stress_score",
        "self_reported_happiness",
    ]
).copy()

print("\nEDA-ready rows (key vars non-missing):", len(df1_eda))
display(df1_eda.head(3))
```

Together, these supporting datasets help us evaluate whether the main Dataset #1 findings are directionally consistent across related but imperfectly matched measures. In the code cell below, the final output table also makes explicit which exposure, active-use, and wellbeing proxies are available in each processed supporting dataset.

```python
from pathlib import Path
import re
import pandas as pd
from IPython.display import display, Markdown

RAW_DIR = Path("data/00-raw")
INTERIM_DIR = Path("data/01-interim")
PROCESSED_DIR = Path("data/02-processed")
EMIR_DIR = RAW_DIR / "emirhanai"
SMMH_LOCAL_PATH = RAW_DIR / "smmh.csv"
SMMH_URL = "https://raw.githubusercontent.com/Daerkns/correlation-between-social-media-and-mental-health/main/smmh.csv"

INTERIM_DIR.mkdir(parents=True, exist_ok=True)
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

summary_rows = []
analysis_rows = []


def snake_case(columns):
    """Convert raw column names to lowercase snake_case labels."""
    return [re.sub(r"[^0-9a-zA-Z]+", "_", str(col)).strip("_").lower() for col in columns]


def numericize(df, columns):
    """Coerce selected columns to numeric values when present."""
    for col in columns:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")


def show_preview(title, df, processed_path, preview_cols):
    """Display a short quality-check summary for a cleaned dataset."""
    display(Markdown(f"#### {title}"))
    print("Processed file:", processed_path)
    print("Shape:", df.shape)
    print("Duplicate rows:", int(df.duplicated().sum()))

    missing_tbl = (
        df[preview_cols]
        .isna()
        .sum()
        .to_frame("missing_count")
        .assign(missing_pct=lambda x: (x["missing_count"] / len(df) * 100).round(2))
    )

    display(Markdown("Preview of cleaned columns"))
    display(df[preview_cols].head(3))
    display(Markdown("Missingness in preview columns"))
    display(missing_tbl)


def parse_platforms(value):
    """Split a comma-separated platform field into normalized unique names."""
    if pd.isna(value):
        return []

    cleaned = []
    for platform in str(value).split(","):
        label = platform.strip().lower()
        if label and label not in cleaned:
            cleaned.append(label)
    return cleaned


# Dataset #2: Mental Health and Social Media Balance

dataset_2_path = RAW_DIR / "Mental_Health_and_Social_Media_Balance_Dataset.csv"
dataset_2_processed = PROCESSED_DIR / "dataset2_social_media_balance_processed.csv"

if dataset_2_path.exists():
    d2 = pd.read_csv(dataset_2_path).copy()
    d2.columns = snake_case(d2.columns)

    numericize(
        d2,
        [
            "age",
            "daily_screen_time_hrs",
            "sleep_quality_1_10",
            "stress_level_1_10",
            "days_without_social_media",
            "exercise_frequency_week",
            "happiness_index_1_10",
        ],
    )

    d2 = d2[d2["age"].notna() & (d2["age"] >= 18)].copy()
    d2 = d2.drop_duplicates().reset_index(drop=True)
    d2["uses_instagram"] = d2["social_media_platform"].astype(str).str.contains("instagram", case=False, na=False)

    d2.to_csv(dataset_2_processed, index=False)

    summary_rows.append(
        {
            "dataset": "Dataset #2",
            "processed_rows": len(d2),
            "processed_columns": d2.shape[1],
            "processed_file": str(dataset_2_processed),
            "notes": "adult-only rows, snake_case columns, duplicate rows removed",
        }
    )
    analysis_rows.append(
        {
            "dataset": "Dataset #2",
            "passive_or_exposure_proxy": "daily_screen_time_hrs",
            "active_proxy": "not directly available",
            "wellbeing_proxy": "stress_level_1_10, happiness_index_1_10, sleep_quality_1_10",
            "processed_file": str(dataset_2_processed),
        }
    )
else:
    d2 = None
    summary_rows.append(
        {
            "dataset": "Dataset #2",
            "processed_rows": "missing",
            "processed_columns": "missing",
            "processed_file": str(dataset_2_processed),
            "notes": "raw file not found locally",
        }
    )


# Dataset #3: Social Media Usage and Emotional Wellbeing

dataset_3_processed = PROCESSED_DIR / "dataset3_social_media_emotion_processed.csv"
d3_frames = []

for split in ["train", "val", "test"]:
    split_path = EMIR_DIR / f"{split}.csv"
    if split_path.exists():
        split_df = pd.read_csv(split_path, engine="python", on_bad_lines="skip")
        split_df["data_split"] = split
        d3_frames.append(split_df)

if d3_frames:
    d3 = pd.concat(d3_frames, ignore_index=True)
    d3 = d3.dropna(how="all").copy()
    d3.columns = snake_case(d3.columns)

    numericize(
        d3,
        [
            "age",
            "daily_usage_time_minutes",
            "posts_per_day",
            "likes_received_per_day",
            "comments_received_per_day",
            "messages_sent_per_day",
        ],
    )

    d3 = d3[d3["age"].notna() & (d3["age"] >= 18)].copy()
    d3 = d3.drop_duplicates().reset_index(drop=True)
    d3["is_instagram"] = d3["platform"].astype(str).str.lower().eq("instagram")
    d3["active_interaction_proxy"] = d3[["posts_per_day", "messages_sent_per_day"]].fillna(0).sum(axis=1)

    d3.to_csv(dataset_3_processed, index=False)

    summary_rows.append(
        {
            "dataset": "Dataset #3",
            "processed_rows": len(d3),
            "processed_columns": d3.shape[1],
            "processed_file": str(dataset_3_processed),
            "notes": "train/val/test combined, blank rows dropped, interaction proxy created",
        }
    )
    analysis_rows.append(
        {
            "dataset": "Dataset #3",
            "passive_or_exposure_proxy": "daily_usage_time_minutes",
            "active_proxy": "posts_per_day, messages_sent_per_day, active_interaction_proxy",
            "wellbeing_proxy": "dominant_emotion",
            "processed_file": str(dataset_3_processed),
        }
    )
else:
    d3 = None
    summary_rows.append(
        {
            "dataset": "Dataset #3",
            "processed_rows": "missing",
            "processed_columns": "missing",
            "processed_file": str(dataset_3_processed),
            "notes": "train/val/test raw files not found locally",
        }
    )


# Dataset #4: Social Media and Mental Health

dataset_4_processed = PROCESSED_DIR / "dataset4_smmh_processed.csv"
dataset_4_source = SMMH_LOCAL_PATH if SMMH_LOCAL_PATH.exists() else SMMH_URL

d4 = pd.read_csv(dataset_4_source).copy()
d4 = d4.rename(
    columns={
        "1. What is your age?": "age",
        "2. Gender": "gender",
        "3. Relationship Status": "relationship_status",
        "4. Occupation Status": "occupation_status",
        "5. What type of organizations are you affiliated with?": "organization_affiliation",
        "6. Do you use social media?": "uses_social_media_answer",
        "7. What social media platforms do you commonly use?": "platforms",
        "8. What is the average time you spend on social media every day?": "avg_time_category",
        "9. How often do you find yourself using Social media without a specific purpose?": "use_without_purpose",
        "10. How often do you get distracted by Social media when you are busy doing something?": "distracted_when_busy",
        "11. Do you feel restless if you haven't used Social media in a while?": "restless_without_social_media",
        "12. On a scale of 1 to 5, how easily distracted are you?": "easy_distractibility_1_5",
        "13. On a scale of 1 to 5, how much are you bothered by worries?": "worry_1_5",
        "14. Do you find it difficult to concentrate on things?": "difficulty_concentrating",
        "15. On a scale of 1-5, how often do you compare yourself to other successful people through the use of social media?": "social_comparison_1_5",
        "16. Following the previous question, how do you feel about these comparisons, generally speaking?": "comparison_feelings",
        "17. How often do you look to seek validation from features of social media?": "validation_seeking",
        "18. How often do you feel depressed or down?": "depressed_or_down",
        "19. On a scale of 1 to 5, how frequently does your interest in daily activities fluctuate?": "interest_fluctuation_1_5",
        "20. On a scale of 1 to 5, how often do you face issues regarding sleep?": "sleep_issues_1_5",
    }
).copy()

numericize(
    d4,
    [
        "age",
        "use_without_purpose",
        "distracted_when_busy",
        "restless_without_social_media",
        "easy_distractibility_1_5",
        "worry_1_5",
        "difficulty_concentrating",
        "social_comparison_1_5",
        "validation_seeking",
        "depressed_or_down",
        "interest_fluctuation_1_5",
        "sleep_issues_1_5",
    ],
)

time_map = {
    "Less than an Hour": 0.5,
    "Between 1 and 2 hours": 1.5,
    "Between 2 and 3 hours": 2.5,
    "Between 3 and 4 hours": 3.5,
    "Between 4 and 5 hours": 4.5,
    "More than 5 hours": 5.5,
}

platform_list = d4["platforms"].map(parse_platforms)
d4["platform_count"] = platform_list.map(len)
d4["platforms_clean"] = platform_list.map(lambda lst: ", ".join(lst) if lst else pd.NA).astype("string")
d4["uses_instagram"] = platform_list.map(lambda lst: "instagram" in lst).astype("boolean")
d4["uses_social_media_original"] = (
    d4["uses_social_media_answer"].astype(str).str.strip().str.lower().map({"yes": True, "no": False}).astype("boolean")
)
d4["avg_time_hours"] = d4["avg_time_category"].map(time_map)

inferred_use = (d4["platform_count"] > 0) | d4["avg_time_hours"].notna()
d4["inconsistent_usage_response"] = (
    (d4["uses_social_media_original"] == False) & inferred_use
).astype("boolean")
d4["uses_social_media"] = (
    d4["uses_social_media_original"].fillna(False).astype("boolean") | inferred_use.astype("boolean")
)

d4 = d4[d4["age"].notna() & (d4["age"] >= 18)].copy()
d4 = d4[d4["uses_social_media"].fillna(False)].copy()
d4 = d4.drop_duplicates().reset_index(drop=True)

d4.to_csv(dataset_4_processed, index=False)

summary_rows.append(
    {
        "dataset": "Dataset #4",
        "processed_rows": len(d4),
        "processed_columns": d4.shape[1],
        "processed_file": str(dataset_4_processed),
        "notes": f"adult social-media users only, parsed platform lists, inconsistency flag added, local raw preferred over mirror ({dataset_4_source})",
    }
)
analysis_rows.append(
    {
        "dataset": "Dataset #4",
        "passive_or_exposure_proxy": "avg_time_hours, use_without_purpose, social_comparison_1_5",
        "active_proxy": "not directly available",
        "wellbeing_proxy": "depressed_or_down, worry_1_5, sleep_issues_1_5",
        "processed_file": str(dataset_4_processed),
    }
)


# Dataset #5: Screen Time vs Mental Wellness

dataset_5_path = RAW_DIR / "ScreenTime vs MentalWellness.csv"
dataset_5_processed = PROCESSED_DIR / "dataset5_screen_time_mental_wellness_processed.csv"

if dataset_5_path.exists():
    d5 = pd.read_csv(dataset_5_path).copy()
    d5 = d5.loc[:, ~d5.columns.str.contains(r"^Unnamed", case=False, na=False)].copy()
    d5.columns = snake_case(d5.columns)

    numericize(
        d5,
        [
            "age",
            "screen_time_hours",
            "work_screen_hours",
            "leisure_screen_hours",
            "sleep_hours",
            "sleep_quality_1_5",
            "stress_level_0_10",
            "productivity_0_100",
            "exercise_minutes_per_week",
            "social_hours_per_week",
            "mental_wellness_index_0_100",
        ],
    )

    d5 = d5[d5["age"].notna() & (d5["age"] >= 18)].copy()
    d5 = d5.drop_duplicates().reset_index(drop=True)

    d5.to_csv(dataset_5_processed, index=False)

    summary_rows.append(
        {
            "dataset": "Dataset #5",
            "processed_rows": len(d5),
            "processed_columns": d5.shape[1],
            "processed_file": str(dataset_5_processed),
            "notes": "adult-only rows, blank unnamed column removed, numeric columns standardized",
        }
    )
    analysis_rows.append(
        {
            "dataset": "Dataset #5",
            "passive_or_exposure_proxy": "screen_time_hours, leisure_screen_hours",
            "active_proxy": "not directly available",
            "wellbeing_proxy": "stress_level_0_10, mental_wellness_index_0_100, sleep_quality_1_5",
            "processed_file": str(dataset_5_processed),
        }
    )
else:
    d5 = None
    summary_rows.append(
        {
            "dataset": "Dataset #5",
            "processed_rows": "missing",
            "processed_columns": "missing",
            "processed_file": str(dataset_5_processed),
            "notes": "raw file not found locally",
        }
    )


supporting_dataset_summary = pd.DataFrame(summary_rows)
harmonized_dataset_map = pd.DataFrame(analysis_rows)

display(Markdown("### Supporting datasets wrangled and saved"))
display(supporting_dataset_summary)
display(Markdown("### Harmonized variables available for later supporting analyses"))
display(harmonized_dataset_map)

if d2 is not None:
    show_preview(
        "Dataset #2 cleaned preview",
        d2,
        dataset_2_processed,
        [
            "age",
            "daily_screen_time_hrs",
            "stress_level_1_10",
            "happiness_index_1_10",
            "sleep_quality_1_10",
            "uses_instagram",
        ],
    )

if d3 is not None:
    show_preview(
        "Dataset #3 cleaned preview",
        d3,
        dataset_3_processed,
        [
            "age",
            "platform",
            "daily_usage_time_minutes",
            "posts_per_day",
            "messages_sent_per_day",
            "active_interaction_proxy",
            "dominant_emotion",
        ],
    )

show_preview(
    "Dataset #4 cleaned preview",
    d4,
    dataset_4_processed,
    [
        "age",
        "avg_time_hours",
        "platform_count",
        "uses_social_media_original",
        "uses_social_media",
        "inconsistent_usage_response",
        "use_without_purpose",
        "social_comparison_1_5",
        "depressed_or_down",
        "sleep_issues_1_5",
        "uses_instagram",
    ],
)

if d5 is not None:
    show_preview(
        "Dataset #5 cleaned preview",
        d5,
        dataset_5_processed,
        [
            "age",
            "screen_time_hours",
            "leisure_screen_hours",
            "stress_level_0_10",
            "mental_wellness_index_0_100",
            "sleep_quality_1_5",
        ],
    )
```

### Analysis 1: Primary Controlled Associations in Dataset #1

This is the main results section for our project because Dataset #1 is our primary evidence source. We estimate correlational relationships between passive use (`time_on_feed_per_day`) and active use (`posts_created_per_week`) with two outcomes (`perceived_stress_score`, `self_reported_happiness`) while controlling for age, gender, country, urban/rural setting, income, employment, education, exercise, sleep, and weekly work hours.

To compare passive and active proxies on a common scale, we report standardized coefficients and partial correlations (residual associations after controls). These estimates are interpreted as adjusted associations, not causal effects.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.linear_model import LinearRegression
from IPython.display import display, Markdown

# Build the analysis frame with required predictors, outcomes, and controls
num_controls = ["age", "exercise_hours_per_week", "sleep_hours_per_night", "weekly_work_hours"]
cat_controls = ["gender", "country", "urban_rural", "income_level", "employment_status", "education_level"]

analysis_cols = [
    "time_on_feed_per_day",
    "posts_created_per_week",
    "perceived_stress_score",
    "self_reported_happiness",
] + num_controls + cat_controls

analysis1_df = df1_eda[analysis_cols].dropna().copy()

# Subsample for computational tractability while retaining a large N
n_cap = 300000
if len(analysis1_df) > n_cap:
    analysis1_df = analysis1_df.sample(n=n_cap, random_state=108)

for col in num_controls + [
    "time_on_feed_per_day",
    "posts_created_per_week",
    "perceived_stress_score",
    "self_reported_happiness",
]:
    analysis1_df[col] = pd.to_numeric(analysis1_df[col], errors="coerce")

analysis1_df = analysis1_df.dropna().copy()

X_num = analysis1_df[num_controls].astype(float)
X_cat = pd.get_dummies(analysis1_df[cat_controls].astype("category"), drop_first=True)
X_controls = pd.concat([X_num, X_cat], axis=1).astype(float)

predictors = ["time_on_feed_per_day", "posts_created_per_week"]
outcomes = ["perceived_stress_score", "self_reported_happiness"]


def partial_corr(y, x, controls):
    reg = LinearRegression()
    reg.fit(controls, y)
    y_res = y - reg.predict(controls)

    reg.fit(controls, x)
    x_res = x - reg.predict(controls)

    return stats.pearsonr(x_res, y_res)


rows = []
for outcome in outcomes:
    y = analysis1_df[outcome].to_numpy()

    # Standardized coefficients from a joint model with both usage proxies + controls
    X_full = pd.concat([analysis1_df[predictors], X_controls], axis=1).astype(float)
    X_std = (X_full - X_full.mean()) / X_full.std(ddof=0).replace(0, 1)

    y_std = (analysis1_df[outcome] - analysis1_df[outcome].mean()) / analysis1_df[outcome].std(ddof=0)

    reg_full = LinearRegression()
    reg_full.fit(X_std, y_std)
    coef_map = dict(zip(X_full.columns, reg_full.coef_))

    for predictor in predictors:
        r_partial, p_partial = partial_corr(
            analysis1_df[outcome].to_numpy(),
            analysis1_df[predictor].to_numpy(),
            X_controls.to_numpy(),
        )
        rows.append(
            {
                "outcome": outcome,
                "predictor": predictor,
                "standardized_beta": float(coef_map[predictor]),
                "partial_r": float(r_partial),
                "partial_p": float(p_partial),
                "n_used": int(len(analysis1_df)),
            }
        )

analysis1_results = pd.DataFrame(rows)
analysis1_results["abs_partial_r"] = analysis1_results["partial_r"].abs()

# Visual summary
plot_tbl = analysis1_results.copy()
plot_tbl["comparison"] = plot_tbl["predictor"] + " -> " + plot_tbl["outcome"]

plt.figure(figsize=(9, 4.5))
sns.barplot(data=plot_tbl, x="comparison", y="standardized_beta", color="#4C72B0")
plt.axhline(0, color="black", linewidth=1)
plt.title("Analysis 1: Standardized coefficients (adjusted associations)")
plt.xticks(rotation=20, ha="right")
plt.ylabel("Standardized beta")
plt.xlabel("")
plt.tight_layout()
plt.show()

display(Markdown("#### Analysis 1 results table (Dataset #1, adjusted)"))
display(
    analysis1_results[
        ["outcome", "predictor", "standardized_beta", "partial_r", "partial_p", "n_used"]
    ].sort_values(["outcome", "predictor"]) 
)

# Nearby interpretation in complete sentences
stress_slice = analysis1_results[analysis1_results["outcome"] == "perceived_stress_score"].copy()
happy_slice = analysis1_results[analysis1_results["outcome"] == "self_reported_happiness"].copy()

stronger_stress = stress_slice.sort_values("abs_partial_r", ascending=False).iloc[0]
stronger_happy = happy_slice.sort_values("abs_partial_r", ascending=False).iloc[0]

print(
    f"After adjusting for demographic and lifestyle controls, the strongest stress association is "
    f"{stronger_stress['predictor']} (partial r={stronger_stress['partial_r']:.3f})."
)
print(
    f"For happiness, the strongest adjusted association is {stronger_happy['predictor']} "
    f"(partial r={stronger_happy['partial_r']:.3f})."
)
print(
    "Overall, passive feed time remains directionally more adverse for wellbeing than active posting, "
    "which is consistent with our hypothesis but still interpreted as correlation rather than causation."
)
```

### Analysis 2: Supporting Robustness Checks Across Secondary Datasets

This section tests whether the primary Dataset #1 pattern is directionally consistent in supporting datasets without treating them as equivalent substitutes. We prioritize Dataset #4 first (closest passive/exposure proxies and mental-health-related outcomes), then use Datasets #2 and #5 as additional robustness checks, and use Dataset #3 only as light descriptive context.

Because variable definitions differ across datasets, we compare direction and relative consistency of associations rather than pooling rows or claiming a shared effect size. All findings remain correlational.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from IPython.display import display, Markdown

support_rows = []


def add_spearman_rows(df, x_col, y_cols, dataset_label):
    rows = []
    for y_col in y_cols:
        subset = df[[x_col, y_col]].dropna().copy()
        if len(subset) < 30:
            continue
        rho, pval = stats.spearmanr(subset[x_col], subset[y_col])
        rows.append(
            {
                "dataset": dataset_label,
                "exposure_proxy": x_col,
                "outcome_proxy": y_col,
                "spearman_rho": float(rho),
                "p_value": float(pval),
                "n_used": int(len(subset)),
            }
        )
    return rows


# Dataset #4 (strongest support)
if "d4" in globals() and d4 is not None:
    support_rows.extend(
        add_spearman_rows(
            d4,
            "avg_time_hours",
            ["depressed_or_down", "worry_1_5", "sleep_issues_1_5"],
            "Dataset #4",
        )
    )
    support_rows.extend(
        add_spearman_rows(
            d4,
            "social_comparison_1_5",
            ["depressed_or_down", "worry_1_5", "sleep_issues_1_5"],
            "Dataset #4",
        )
    )

# Dataset #2
if "d2" in globals() and d2 is not None:
    support_rows.extend(
        add_spearman_rows(
            d2,
            "daily_screen_time_hrs",
            ["stress_level_1_10", "happiness_index_1_10", "sleep_quality_1_10"],
            "Dataset #2",
        )
    )

# Dataset #5
if "d5" in globals() and d5 is not None:
    support_rows.extend(
        add_spearman_rows(
            d5,
            "leisure_screen_hours",
            ["stress_level_0_10", "mental_wellness_index_0_100", "sleep_quality_1_5"],
            "Dataset #5",
        )
    )

support_tbl = pd.DataFrame(support_rows)

if support_tbl.empty:
    print("No supporting-dataset rows available for Analysis 2.")
else:
    display(Markdown("#### Analysis 2 summary table (supporting datasets)"))
    display(support_tbl.sort_values(["dataset", "exposure_proxy", "outcome_proxy"]))

    # Heatmap for quick directional consistency check
    heat_df = support_tbl.copy()
    heat_df["pair"] = heat_df["dataset"] + ": " + heat_df["exposure_proxy"] + " -> " + heat_df["outcome_proxy"]

    plt.figure(figsize=(8, max(4, 0.35 * len(heat_df))))
    sns.heatmap(
        heat_df[["spearman_rho"]],
        annot=np.array(heat_df["spearman_rho"].round(3)).reshape(-1, 1),
        fmt="",
        cmap="coolwarm",
        center=0,
        yticklabels=heat_df["pair"],
        cbar_kws={"label": "Spearman rho"},
    )
    plt.title("Analysis 2: Direction and consistency across supporting datasets")
    plt.xticks([0.5], ["rho"])
    plt.tight_layout()
    plt.show()

    # Dataset #3 light descriptive context only
    if "d3" in globals() and d3 is not None:
        d3_context = d3[["daily_usage_time_minutes", "dominant_emotion", "active_interaction_proxy"]].dropna().copy()

        usage_q = pd.qcut(
            d3_context["daily_usage_time_minutes"],
            q=4,
            labels=["Q1-low", "Q2", "Q3", "Q4-high"],
            duplicates="drop",
        )
        emotion_mix = (
            d3_context.assign(usage_quartile=usage_q)
            .groupby(["usage_quartile", "dominant_emotion"], as_index=False)
            .size()
            .rename(columns={"size": "count"})
        )
        emotion_mix["share"] = emotion_mix["count"] / emotion_mix.groupby("usage_quartile")["count"].transform("sum")
        emotion_mix = emotion_mix.sort_values(["usage_quartile", "share"], ascending=[True, False])

        display(Markdown("#### Dataset #3 descriptive context (not a primary outcome test)"))
        display(emotion_mix.head(12))

    # Nearby interpretation in full sentences
    stress_like = support_tbl[support_tbl["outcome_proxy"].str.contains("stress|worry|depressed", case=False, na=False)]
    positive_stress_links = int((stress_like["spearman_rho"] > 0).sum())

    wellness_like = support_tbl[support_tbl["outcome_proxy"].str.contains("happiness|wellness|sleep_quality", case=False, na=False)]
    adverse_wellness_links = int((wellness_like["spearman_rho"] < 0).sum())

    print(
        "Across supporting datasets, higher passive/exposure-style use is generally associated with worse stress-like outcomes "
        f"in {positive_stress_links} tested pairs and with lower positive-wellbeing proxies in {adverse_wellness_links} pairs."
    )
    print(
        "These secondary results support the direction of Analysis 1 without replacing Dataset #1 as the primary evidence source."
    )
```

## Ethics

Our analysis uses observational, mostly self-reported survey-style data, so ethical reporting requires careful limits on interpretation. We therefore frame all results as associations rather than causal effects. Even when patterns are strong, they may still reflect confounding from unmeasured factors such as baseline mental health, offline social support, or life stressors not captured in the available variables.

A second concern is measurement quality. The primary dataset and several supporting datasets rely on self-reported exposure and wellbeing proxies, which can introduce recall error and social desirability bias. We also use proxy variables to represent complex constructs (for example, passive use via `time_on_feed_per_day` and active use via `posts_created_per_week`), which may not capture all relevant behavior dimensions.

A third concern is representation and external validity. Some supporting datasets have relatively small sample sizes and may not represent all age groups, countries, or socioeconomic contexts. In addition, near-zero missingness in Dataset #1 raises the possibility that data may be synthetic or heavily processed, which can limit real-world generalizability even when internal patterns are statistically clear.

A fourth concern is privacy and data handling. Although the project uses publicly available files and no direct identifiers in the final processed tables, responsible practice still requires minimizing unnecessary personal-level exposure, avoiding attempts at re-identification, and reporting only aggregated findings.

Given these issues, we explicitly avoid overclaiming. Our goal is to provide transparent, reproducible, correlational evidence about how passive and active social-media behaviors relate to adult wellbeing, while acknowledging that causal inference would require stronger designs such as longitudinal tracking or randomized interventions.

## Discussion and Conclusion

The main finding of this project is that passive social-media exposure shows a more consistently adverse association with adult wellbeing than active posting behavior. In the primary Dataset #1 analysis, higher `time_on_feed_per_day` aligns with higher stress and lower happiness, while `posts_created_per_week` shows weaker and more mixed adjusted associations. This pattern remains visible in exploratory plots and in the controlled models that account for demographic, socioeconomic, and lifestyle covariates.

The supporting analyses reinforce, rather than replace, this conclusion. Dataset #4 (our strongest supporting source) shows that passive/exposure-related variables such as time spent and social comparison tend to move in the adverse direction with depressed mood, worry, and sleep issues. Datasets #2 and #5 provide additional robustness checks where broader exposure measures are generally directionally consistent with higher stress-like outcomes and weaker wellbeing proxies. Dataset #3 is used only as descriptive context because `dominant_emotion` is not equivalent to our primary stress and happiness outcomes.

Relative to prior work, our results are broadly compatible with literature arguing that context and behavior type matter more than simple total-use narratives. We do not find evidence that active behavior is uniformly beneficial across outcomes, and we also do not claim that passive behavior directly causes poorer wellbeing. Instead, our results suggest that passive-heavy usage patterns are a more reliable correlate of adverse wellbeing in this multi-dataset setting.

The findings should be interpreted with clear limitations. All datasets are observational and mostly self-reported, measurement proxies are imperfect, and some datasets are small or context-specific. Dataset #1 may also be synthetic or heavily processed, which can affect external validity even with strong internal consistency. Because of these constraints, we treat all estimates as correlational and avoid policy-style causal claims.

If this work were extended, the strongest next steps would be (1) longitudinal data to test temporal ordering, (2) richer behavioral measures that better separate passive and active use dimensions, and (3) outcome measures beyond self-report alone. Within the current project scope, the evidence supports a cautious but clear takeaway: among adults in these datasets, passive social-media exposure is more consistently associated with worse wellbeing than active posting behavior, while uncertainty remains about mechanism and causality.
